```python
folder = 'C:\\Users\\jessi\\Documents\\TCC Docs\\Datasets Covid\\'
```

```python
#Dados Sírio Líbanes
hsl_pacientes = pd.read_csv(folder + "HSL_Agosto\\final\\HSL_Pacientes_2.csv", sep = "|", encoding="utf-8")
hsl_exames = pd.read_csv(folder + "HSL_Agosto\\final\\HSL_Exames_2.csv", sep = "|", encoding="utf-8")
hsl_exames['DE_EXAME'] = hsl_exames['DE_EXAME'] + " - " + hsl_exames['DE_ANALITO']
hsl_exames.shape
```

```python
df = pd.DataFrame()
df['EXAME'] = hsl_exames['DE_EXAME']
df['UNIDADE'] = hsl_exames['CD_UNIDADE']
df['VALOR_REF'] = hsl_exames['DE_VALOR_REFERENCIA']
technical_hsl = df.drop_duplicates()
technical_hsl.to_csv('HSL_Exames.csv', index=False, sep="|",encoding='utf-8-sig')
```

```python
cols_exam = hsl_exames['DE_EXAME'].value_counts().to_frame(name='Preenchidos')
cols_exam['%'] = (cols_exam['Preenchidos'] / hsl_exames.shape[0]) * 100
to_drop = cols_exam[cols_exam['Preenchidos'] < 500].sort_values(by='%', ascending=False).index.to_list()
hsl_exames = hsl_exames[~hsl_exames['DE_EXAME'].isin(to_drop)]
hsl_exames['DE_EXAME'].value_counts()
```

```python
hsl_exames['DT_COLETA'] = pd.to_datetime(hsl_exames['DT_COLETA'], format='%Y/%m/%d').dt.strftime('%d/%m/%Y')
to_drop = ['*', 'NOVA COLETA', 'Nova coleta', 'nova coleta', 'Nova Coleta', 'Indeterminado', 'Inconclusivo']
hsl_exames = hsl_exames[~hsl_exames['DE_RESULTADO'].isin(to_drop)]
del hsl_exames['DE_ANALITO']
del hsl_exames['ID_ATENDIMENTO']
hsl_exames.sort_values(['ID_PACIENTE','DT_COLETA', 'DE_EXAME'], ascending=[False, False, False],inplace=True)
```

```python
hsl_exames_exp = hsl_exames.pivot_table(index= ['ID_PACIENTE','DT_COLETA'], columns='DE_EXAME', values='DE_RESULTADO', aggfunc=';'.join)
```

```python
cols_hsl = pd.DataFrame(hsl_exames_exp.notna().sum(),columns=['Preenchidos'])
cols_hsl['%'] = (cols_hsl['Preenchidos'] / hsl_exames_exp.shape[0]) * 100
cols_hsl[cols_hsl.index.str.contains('Bilirrubinas')].sort_values(by='%', ascending=False)
```

```python
cols_to_save = ['Bilirrubinas Total E Fracoes - Bilirrubina Direta', 'Bilirrubinas Total E Fracoes - Bilirrubina Indireta',
                'Bilirrubinas Total E Fracoes - Bilirrubina Total', 'Creatinina - Creatinina', 
                'Fosfatase Alcalina - Fosfatase Alcalina',
                'Gama Gt - Gama-GT', 'Glicemia, plasma - Glicose', 'Magnesio - Magnésio',
                'Potássio - Potássio', 'Sódio - Sódio', 'Tgo - AST (TGO)', 'Tgp - ALT (TGP)', 'Hemograma - Basófilos (%)',
                'Hemograma - Basófilos', 'Hemograma - Eosinófilos (%)', 'Hemograma - Eosinófilos', 'Hemograma - Linfócitos (%)',
                'Hemograma - Linfócitos', 'Hemograma - Monócitos (%)', 'Hemograma - Monócitos', 'Hemograma - Neutrófilos (%)',
                'Hemograma - Neutrófilos', 'Hemograma - CHCM', 'Hemograma - HCM', 'Hemograma - Hematócrito',
                'Hemograma - Hemoglobina', 'Hemograma - Leucócitos', 'Hemograma - Plaquetas', 'Hemograma - RDW',
                'Hemograma - Eritrócitos', 'Hemograma - Volume plaquetário médio',
                'COVID-19-PCR para SARS-COV-2, Vários Materiais (Fleury) - Coronavírus (2019-nCoV)',
                'Urina Tipo I - Jato Medio - Aspecto, urina', 'Urina Tipo I - Jato Medio - Glicose, urina isol',
                'Urina Tipo I - Jato Medio - Cilindros, urina', 'Urina Tipo I - Jato Medio - Cor, urina', 
                'Urina Tipo I - Jato Medio - C.Cetônicos, urina', 'Urina Tipo I - Jato Medio - Densidade, urina',
                'Urina Tipo I - Jato Medio - Leucócitos, urina', 'Urina Tipo I - Jato Medio - Nitrito, urina',
                'Urina Tipo I - Jato Medio - Proteínas, urina isolada', 'Urina Tipo I - Jato Medio - Urobilinogênio,urina',
                'Urina Tipo I - Jato Medio - pH, urina']
```

```python
df_hsl_exames = hsl_exames_exp[cols_to_save]
```

```python
cols_hemograma = cols_hsl[cols_hsl.index.str.contains('Hemograma')].index
cols_urina = cols_hsl[cols_hsl.index.str.contains('Urina Tipo I')].index
df_hsl_exames = df_hsl_exames[df_hsl_exames[cols_urina[0]].notna() == True]
df_hsl_exames =  df_hsl_exames[df_hsl_exames['Hemograma - Basófilos'].notna() == True]
```

```python
df_hsl_exames['DIAGNOSTICO_COVID'] = 'PCR - ' + df_hsl_exames['COVID-19-PCR para SARS-COV-2, Vários Materiais (Fleury) - Coronavírus (2019-nCoV)']
df_hsl_exames = df_hsl_exames[~df_hsl_exames['DIAGNOSTICO_COVID'].isna()]
```
