```python
import pandas as pd
import numpy as np
import regex as rg
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df_einstein_exames = pd.read_csv('Einstein_v4.csv', sep = '|', encoding = 'utf-8-sig')
df_fleury_exames = pd.read_csv('Fleury_v4.csv', sep = '|', encoding = 'utf-8-sig')
df_hsl_exames = pd.read_csv('HSL_v4.csv', sep = '|', encoding = 'utf-8-sig')
```

```python
def exclui_outliers(df, col_name):
    intervalo= 2.7*df[col_name].std()
    media = df[col_name].mean()
    df.loc[df[col_name] < (media - intervalo), col_name] = np.nan
    df.loc[df[col_name] > (media + intervalo), col_name] = np.nan
    df.dropna(inplace=True)
```

```python
df_fleury_exames['DIAGNOSTICO_COVID'].value_counts()
```

```python
df_einstein_copy = df_einstein_exames.copy()
cols_num_einstein = df_einstein_exames.select_dtypes('number').columns.to_list()
for col in cols_num_einstein:
    exclui_outliers(df_einstein_copy, col)
print(df_einstein_copy.shape)
print(df_einstein_copy['DIAGNOSTICO_COVID'].value_counts())
df_einstein_copy.iloc[:,:13].boxplot(figsize=(10,7), rot=0, fontsize=15, vert=False)
```

```python
cols_num_fleury = df_fleury_exames.select_dtypes('number').columns.to_list()
z_scores_fleury = stats.zscore(df_fleury_exames[cols_num_fleury])
abs_z_scores_fleury = np.abs(z_scores_fleury)
filtered_entries_fleury = (abs_z_scores_fleury < 3).all(axis=1)
new_df_fleury_exames = df_fleury_exames[filtered_entries_fleury]
print(new_df_fleury_exames.shape)
print(new_df_fleury_exames['DIAGNOSTICO_COVID'].value_counts())
new_df_fleury_exames.iloc[:,:13].boxplot(figsize=(10,7), rot=0, fontsize=15, vert=False)
```
