```python
def is_number(n_str):
    if ',' in n_str:
        n_str = n_str.replace(',','.')
    try:
        float(n_str)
        return True
    except ValueError:
        return False
    
def new_value(input_str):
    list_val = input_str.split(";")
    
    if is_number(list_val[0]): #Se forem números
        float_list = [float(i.replace(',','.')) for i in list_val]

        if 0.0 in float_list:
            float_list.remove(0.)
            if not float_list: #Se a lista só tinha 0's e ficou vazia
                return 0

        if len(float_list) != len(set(float_list)) or len(float_list) == 1: #Se houverem valores repetidos, retorna a Moda
            return stats.mode(float_list)[0][0]
        else: #Senão, retorna a mediana
            return np.median(np.asarray(float_list))

    else: #Se forem variáveis categóricas
        return stats.mode(list_val)[0][0] #Retorna valor mais frequente         
```
